```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1) #entre para está, respectivamente, o numero de linhas, colunas e a posição da figura 
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados['Valor'], dados['Area'])#grafico de dispersão
g1.set_title('Valor x àrea')# Define o titulo do gráfico

g2.hist(dados['Valor']) # Hitograma
g2.set_title('Histograma')

dados_g3 = dados['Valor'].sample(100) #retira sao amostras aleatórias
dados_g3.index = range(dados_g3.shape[0]) #Cria um index com base no numero de linhas
g3.plot(dados_g3)
g3.set_title('Amostra de Valor')

grupo = dados.groupby('Tipo')['Valor'] #Agrupo pela coluna tipo, em relação a coluna valor
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor médio por tipo')
```
